# Sorting algorithms

```python
import numpy as np 
import matplotlib.pyplot as plt 

np.random.seed(0)

def generate_n_random_numbers(n):
    return np.random.rand(n)

def plot_bar(y):
    x = np.arange(len(y))
    plt.bar(x, y)
    plt.grid()
    plt.show()
```

```python
random_numbers = generate_n_random_numbers(11)
plot_bar(random_numbers)
```

## Bubble Sort

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

sorted_numbers = random_numbers.copy()
bubble_sort(sorted_numbers)
plot_bar(sorted_numbers)
```

# Quick Sort

```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quicksort(left) + middle + quicksort(right)

sorted_numbers = quicksort(random_numbers)
plot_bar(sorted_numbers)
```

# Merge sort

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]
    
    left = merge_sort(left)
    right = merge_sort(right)
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
    while i < len(left):
        result.append(left[i])
        i += 1
    
    while j < len(right):
        result.append(right[j])
        j += 1
    
    return result

sorted_numbers = merge_sort(random_numbers)
plot_bar(sorted_numbers)
```
