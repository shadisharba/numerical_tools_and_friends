## Random plot

A plot of the following variables: 

$$ y_1 = sin(x) $$ 
$$ y_2 = cos(x) $$ 

```python
x  = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)
fig = plt.figure()
plt.plot(x,y1,'b',label=r'$y_1 = sin(x)$')
plt.plot(x,y2,'r',label=r'$y_2 = cos(x)$')
plt.xlabel(r'$x$ [-]')
plt.ylabel(r'$y_1, y_2$ [-]')
plt.title('A plot')
plt.legend()
plt.grid('on')
```

## Parametrized plot

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

@interact(a=(0, 10, 0.1))
def plot_parametrized_figure(a):
    x = np.linspace(0, 2 * np.pi, 100)
    y = np.sin(a * x)
    
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'Parametrized Figure (a = {a})')
    ax.grid(True)
    plt.show()
```
